# Q-Q(Quantile-Quantile) Plot

- If the distribution is Normal/Gaussian we can get many information from plot e.g. 68–95–99.7 rule(https://en.wikipedia.org/wiki/68%E2%80%9395%E2%80%9399.7_rule)
- Here we are trying to see if a distrbution is Normal/Guassian or not. One method is using Q-Q plot

### Algorithm:

   Say we have array A of integer(size m)[x1, x2, x3......xm]

- Sort the array and find the percentiles
- Create an array of integer which is normally distributed and plot against array A. If the its a straight line then array A has has Normal distriution.

```python
#Q-Q plot
import numpy as np 
import pylab 
import scipy.stats as stats

# N(0,1)
std_normal = np.random.normal(loc = 0, scale = 1, size=1000)

#loc is mean and scale is standerd deviation

# 0 to 100th percentiles of std-normal
for i in range(0,101):
    print(i, np.percentile(std_normal,i))
#std_normal.sort()
#print(std_normal)
```

```python
# generate 100 sanples from N(20,5)
# Measurements has normal distriution
measurements = np.random.normal(loc = 20, scale = 5, size=40000) 
#try size=1000


stats.probplot(measurements, dist="norm", plot=pylab)
pylab.show()
```

##### We are getting a straight line 

```python
# generate 100 sanples from N(20,5)
# Here measurements has uniform distribution and we should not get straight line
measurements = np.random.uniform(low=-1, high=1, size=10000) 
#try size=1000

stats.probplot(measurements, dist="norm", plot=pylab)
 
pylab.show()
```
